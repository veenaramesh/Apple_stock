## STEP TWO: define dickey-fuller test
[The Dickey-Fuller Test](https://en.wikipedia.org/wiki/Dickey%E2%80%93Fuller_test) tests the null hypothesis that a unit root is present in an autoregressive model; it tests stationarity. 

```python
def dickey_fuller(ts): 
    print("Results of Dickey-Fuller Test: ")
    df = adfuller(ts, autolag="AIC")
    output = pd.Series(df[0:4], index=["test statistic", "p-value", "# lags used", "number of observations used"])
    for percent, value in df[4].items(): 
        output["Critical Value (%s)"%percent] = value
    print(output)
```

```python
dickey_fuller(app)
```

The test statistic is greater than the critical value ($ -2.35 > -3.44$), meaning that the data cannot be considered stationary.
